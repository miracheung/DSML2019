```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Project_Data_1.csv')
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:, 1:]
y = df['Sales of Wheat in tons']
```

```python
X.head()
```

```python
X['1990'] = X['1990'].astype('int64')
```

```python
for i in range(len(X)):
    for j in range(len(X.columns)):
        X.iloc[i,j] = X.iloc[i,j].replace(',', '')
```

```python
X_1 = X.iloc[:, :6]
X_1.head()
```

```python
X_2 = X.iloc[:, 6:14]
X_2.head()
```

```python
X_3 = X.iloc[:, 14:]
X_3.head()
```

```python
for i in range(len(X_1)):
    for j in range(len(X_1.columns)):
        X_1.iloc[i,j] = X_1.iloc[i,j].replace(',', '')
```

```python
for i in range(len(X_3)):
    for j in range(len(X_3.columns)):
        X_3.iloc[i,j] = X_3.iloc[i,j].replace(',', '')
```

```python
X = pd.concat([X_1, X_1, X_3], axis=1)
X.head()
```

```python
X.info()
```

```python
X = X.astype('int64')
```

```python
X.info()
```

```python
from sklearn.preprocessing import StandardScaler

X_std = StandardScaler().fit_transform(X)
X_std
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
clf = pca.fit_transform(X_std)
pca.explained_variance_ratio_.cumsum()
```

```python
pca = PCA(n_components=3)
clf = pca.fit_transform(X_std)
pca.explained_variance_ratio_.cumsum()
```

```python
pca = PCA(n_components=10)
clf = pca.fit_transform(X_std)
pca.explained_variance_ratio_.cumsum()
y = list(pca.explained_variance_ratio_)
y
```

```python
plt.plot(list(range(1, 11)), y)
```

```python
from sklearn.cluster import KMeans

sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(X)
    df["clusters"] = kmeans.labels_
    sse[k] = kmeans.inertia_
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(n_clusters=3).fit(X)
y_kmeans = clustering.labels_

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
g0 = []
g1 = []
g2 = []

for i in range(len(y_kmeans)):
    if y_kmeans[i] == 0:
        g0.append(i)
    elif y_kmeans[i] == 1:
        g1.append(i)
    else:
        g2.append(i)
```
