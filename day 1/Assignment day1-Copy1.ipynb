```python
dfp5 = pd.read_csv('SFO Public Department.csv', encoding='utf-8')
dfp5.head()
```

```python
dfp5_1 = dfp5[['TotalPay', 'Year']]
dfp5_1 = dfp5_1.groupby('Year').sum()
plt.plot(dfp5_1)
```

2

```python
dfp5_2 = dfp5[['JobTitle', 'TotalPay', 'Year']]
groups = dfp5_2.groupby('Year')
dfp5_2 = groups.get_group(2018)
dfp5_2 = dfp5_2.groupby('JobTitle').mean()
dfp5_2.sort_values(by='TotalPay', ascending=False).iloc[0]
```

3

```python
dfp5_3 = dfp5[['OvertimePay', 'Year']]
groups = dfp5_3.groupby('Year')
dfp5_3 = groups.get_group(2018)
dfp5_3.groupby('Year').sum().iloc[0, 0]
```

4

```python
dfp5_4 = dfp5[['JobTitle', 'TotalPayBenefits', 'Year']]
groups = dfp5_4.groupby('Year')
dfp5_4 = groups.get_group(2018)
dfp5_4 = dfp5_4[['JobTitle', 'TotalPayBenefits']]
temp = dfp5_4['JobTitle'].value_counts()
temp = temp.head(5)
lst = list(temp.index)
dfp5_4_1 = dfp5_4.groupby('JobTitle').sum()
print(lst[0], '\t', '\t', dfp5_4_1.get_value(lst[0], 'TotalPayBenefits'))
print(lst[1], '\t', '\t', '\t', dfp5_4_1.get_value(lst[1], 'TotalPayBenefits'))
print(lst[2], '\t', '\t', dfp5_4_1.get_value(lst[2], 'TotalPayBenefits'))
print(lst[3], '\t', dfp5_4_1.get_value(lst[3], 'TotalPayBenefits'))
print(lst[4], '\t', '\t', '\t', dfp5_4_1.get_value(lst[4], 'TotalPayBenefits'))
```
